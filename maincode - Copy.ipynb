```python
# Convert new individual notebooks to HTML exports for embedding
import nbformat
from nbconvert import HTMLExporter

exporter = HTMLExporter()
exporter.exclude_input = True  # hide code cells, only keep markdown + outputs

paths = {
    "ICT305_H1.ipynb": "H1.html",
    "ICT305_A2(H1-H2).ipynb": "H2.html",
    "H3.ipynb": "H3.html"
}

outputs = []
for src, dst in paths.items():
    with open(src, "r", encoding="utf-8") as f:
        nb = nbformat.read(f, as_version=4)
    (body, _) = exporter.from_notebook_node(nb)
    with open(dst, "w", encoding="utf-8") as f:
        f.write(body)
    outputs.append(dst)

outputs
```
